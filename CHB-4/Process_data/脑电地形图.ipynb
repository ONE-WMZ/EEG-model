```python
import mne
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 创建原始电极信息（10-20系统）
ch_names = ['F7', 'T7', 'P7', 'F8', 'T8', 'P8', 'Fz', 'Cz', 'Pz']
info = mne.create_info(ch_names=ch_names, sfreq=200., ch_types='eeg')

# 设置标准10-20蒙太奇
montage = mne.channels.make_standard_montage('standard_1020')
info.set_montage(montage)
```

```python
# 2. 定义双极导联对
bipolar_pairs = [
    ('F7', 'T7'),
    ('T7', 'P7'),
    ('F8', 'T8'),
    ('T8', 'P8'),
    ('Fz', 'Cz'),
    ('Cz', 'Pz')
]
```

```python
# 仅查看电极位置
fig = mne.viz.plot_sensors(info, show_names=True, block=True)
```
