```python
import numpy as np
import matplotlib.pyplot as plt

# 参数设置
sampling_rate = 1000  # 采样率 (Hz)
duration = 2          # 信号持续时间 (秒)
frequency = 5         # 正弦波频率 (Hz)

# 生成时间数组
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# 生成正弦波信号
signal = np.sin(2 * np.pi * frequency * t)+np.sin(2 * np.pi * 50 * t)+np.sin(2 * np.pi * 150 * t)

# 进行傅里叶变换
fourier_transform = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(t), d=1/sampling_rate)

# 只取一半的频谱（因为傅里叶变换是对称的）
positive_frequencies = frequencies[:len(frequencies)//2]
positive_fourier_transform = fourier_transform[:len(fourier_transform)//2]

# 计算幅度谱
magnitude_spectrum = np.abs(positive_fourier_transform)

# 绘制原始信号
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

# 绘制频谱图
plt.subplot(2, 1, 2)
plt.stem(positive_frequencies, magnitude_spectrum, 'b', markerfmt=" ", basefmt="-b")
plt.title('Magnitude Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```
