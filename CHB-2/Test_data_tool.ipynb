```python
import data_tool as tool
import matplotlib.pyplot as plt
```

```python
path = tool.root_path+tool.file_list[5]
data_before = tool.before(path)
data_after = tool.after_edf(path)
```

```python
print(data_after.shape)
```

```python
# 假设 data_before 和 data_after 是形状为 (n_channels, n_samples) 的NumPy数组
plt.figure(figsize=(12, 6))  # 设置图像大小
# 绘制曲线并添加标签
plt.plot(data_before[0, 0:1000], 'b-', label='Before') 
plt.plot(data_after[0, 0:1000], 'r', label='After 1e6')   
# 添加图例和标签
plt.legend(fontsize=12, loc='upper right')  # 显示图例
plt.xlabel('Time (samples)', fontsize=12)   # X轴标签
plt.title('EEG Signal Comparison: Before vs After Processing', fontsize=14)  # 标题

plt.show()
```
